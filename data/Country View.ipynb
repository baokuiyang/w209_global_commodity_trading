# Create CSV for Commodity Details

```python
data = pd.read_csv('all_merged_cl_ix_id_only.csv')
data.head()
```

```python
# Update rtCode for mapping
# Need to update France 251 to 250, Italy 381 to 380, Norway 579 to 578, India 699 to 356, 
# Sudan 729 to 728, Swuzerkabd 757 to 756, USA 842 to 840
data.rtCode[data.rtCode == 251] = 250
data.rtCode[data.rtCode == 381] = 380
data.rtCode[data.rtCode == 579] = 578
data.rtCode[data.rtCode == 699] = 356
data.rtCode[data.rtCode == 729] = 728
data.rtCode[data.rtCode == 757] = 756
data.rtCode[data.rtCode == 842] = 840
```

```python
# Remove useless data
data = data[data.rgCode != 3]
# Remove parent country = 0 since we need detailed values
data = data[data.ptCode != 0]
data.head(10)
```

```python
data.shape
```

```python
# Remove ptCode and Aggregate by yr, rgCode, rtCode, cmdCode
data = data.drop(columns =['ptCode']).groupby(['yr','rgCode','rtCode','cmdCode'], as_index = False).sum()
```

```python
# Join the country names
global_data = pd.merge(left = data, right = df_cn, left_on='rtCode', right_on='Numeric code')\
                .drop(columns=['rtCode', 'Alpha-2 code', 'Numeric code'])
global_data.head()
```

```python
global_data.shape
```

```python
# Get the commodity list
sample_data = pd.read_csv("sample_data_1000_records_id_name.csv")
Commodity = pd.pivot_table(sample_data, index = ['cmdCode','cmdDescE'])\
            .drop(columns = ['TradeValue', 'ptCode', 'rgCode', 'rtCode', 'yr'])\
.reset_index()
```

```python
# Only keep the first one
Commodity = Commodity.drop_duplicates(subset = 'cmdCode', keep = "first")
```

```python
# Join the Commodity list
global_data = pd.merge(left = global_data, right = Commodity, left_on='cmdCode', right_on='cmdCode')
global_data.head()
```

```python
# Add import or export
global_data['trade'] = global_data.apply(lambda row: 'Import' if (row.rgCode == 1) else 'Export', axis = 1).drop(columns=['rgCode'])
```
