```python
import pandas as pd
import numpy as np
```

```python
# Raw Data
df = pd.read_csv('all_merged_cl_ix_id_only.csv')
df.head()
```

# Simple EDA

```python
# Check Report Country
country=sorted(df['rtCode'].unique())
print(country)
print("Total Report Country:", len(country))
```

```python
# Check Parent Country
country_p=sorted(df['ptCode'].unique())
print(country_p)
print("Total Parent Country:", len(country_p))
```

```python
# Remove useless data
df1 = df[df.rgCode != 3]
# Only keep parent country to be 0 since we only need the total value
df1 = df1[df.ptCode == 0]
df1.head(10)
```

```python
# For Mapping purposes
# Need to update France 251 to 250, Italy 381 to 380, Norway 579 to 578, India 699 to 356, 
# Sudan 729 to 728, Swuzerkabd 757 to 756, USA 842 to 840
df1.rtCode[df.rtCode == 251] = 250
df1.rtCode[df.rtCode == 381] = 380
df1.rtCode[df.rtCode == 579] = 578
df1.rtCode[df.rtCode == 699] = 356
df1.rtCode[df.rtCode == 729] = 728
df1.rtCode[df.rtCode == 757] = 756
df1.rtCode[df.rtCode == 842] = 840
```

```python
df1[df1.rtCode== 840].head() # Comfirmed
```

```python
# Aggregate on year, import/export, country code and parent country = 0 (world)
group_multi = df1.groupby(['yr','rgCode','ptCode','rtCode'], as_index = False).sum()
```

```python
# Join Country names
df_cn = pd.read_csv('countries_codes_and_coordinates.csv')
global_view = pd.merge(left = group_multi, right = df_cn, left_on='rtCode', right_on='Numeric code')\
                .drop(columns=['rtCode', 'cmdCode','ptCode','Alpha-2 code', 'Numeric code'])
```

```python
global_view.head()
```

```python
# Add import or export
global_view['trade'] = global_view.apply(lambda row: 'Import' if (row.rgCode == 1) else 'Export', axis = 1)
```

```python
global_view = global_view.drop(columns=['rgCode'])
```

```python
global_view[global_view.Country == "United States"].head()
```

```python
# Save data for global view
global_view.to_csv('Global_view.csv', index=False)
```
